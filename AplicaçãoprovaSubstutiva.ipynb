```python
from ProvaSUB.Repositorio import (paran, karr, model, simu)
```

```python
import matplotlib.pyplot as plt
```

```python
t1, alpha1=simu(20, 298.15, paran())
t2, alpha2=simu(20, 308.15, paran())
t3, alpha3=simu(20, 318.15, paran())
plt.plot(t1, alpha1, label='298.15 K')
plt.plot(t2, alpha2, label='308.15 K')
plt.plot(t3, alpha3, label='318.15 K')
plt.title("Observação da cura de compósitos poliméricos variando com a tempertaura")
plt.xlabel("Tempo (h)")
plt.ylabel("Fator de cura")
plt.legend()
plt.show()
```

Observa-se que ao aumentar a temperatura, o fator de cura aumenta significamente, pois ao aumentar a temperatura, aumenta a troca termica, secando mais rapidamente o compósito polimérico.
O efeito da Temperatura impacta desde o inicio, na constante de arrhenius, o aumento da temperatura é inversamente proporcional ao valor da constante, a fazendo reduzir.
Ao ser inserido no método diferencial, observa-se que irá formar curvas com o decorrer do tempo
